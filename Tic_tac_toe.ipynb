```python
import pandas as pd
import numpy as np
from sklearn import metrics

df = pd.read_csv('tic_tac_toe.csv',index_col=0).iloc[:,:]
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df['Top_left']=le.fit_transform(df['Top_left'])
df['Top_middle']=le.fit_transform(df['Top_middle'])
df['Top_right']=le.fit_transform(df['Top_right'])
df['Middle_left']=le.fit_transform(df['Middle_left'])
df['Middle_middle']=le.fit_transform(df['Middle_middle'])
df['Middle_right']=le.fit_transform(df['Middle_right'])
df['Bottom_left']=le.fit_transform(df['Bottom_left'])
df['Bottom_middle']=le.fit_transform(df['Bottom_middle'])
df['Bottom_right']=le.fit_transform(df['Bottom_right'])
 
print (df)
```

```python
from sklearn.preprocessing import OneHotEncoder 
onehotencoder = OneHotEncoder(sparse=False)  
re=df['Top_left'].values.reshape(len(df['Top_left']),1)
TL= onehotencoder.fit_transform(re)
re=df['Top_middle'].values.reshape(len(df['Top_middle']),1)
TM= onehotencoder.fit_transform(re)
re=df['Top_right'].values.reshape(len(df['Top_right']),1)
TR= onehotencoder.fit_transform(re)
re=df['Middle_left'].values.reshape(len(df['Middle_left']),1)
ML= onehotencoder.fit_transform(re)
re=df['Middle_middle'].values.reshape(len(df['Middle_middle']),1)
MM= onehotencoder.fit_transform(re)
re=df['Middle_right'].values.reshape(len(df['Middle_right']),1)
MR= onehotencoder.fit_transform(re)
re=df['Bottom_left'].values.reshape(len(df['Bottom_left']),1)
BL= onehotencoder.fit_transform(re)
re=df['Bottom_middle'].values.reshape(len(df['Bottom_middle']),1)
BM= onehotencoder.fit_transform(re)
re=df['Bottom_right'].values.reshape(len(df['Bottom_right']),1)
BR= onehotencoder.fit_transform(re)


print(np.shape(TL))
print(np.shape(TM))
print(np.shape(TR))
print(np.shape(ML))
print(np.shape(MM))
print(np.shape(MR))
print(np.shape(BL))
print(np.shape(BM))
print(np.shape(BR))

df
```

```python
from sklearn.model_selection import train_test_split 
from sklearn import model_selection,naive_bayes
from sklearn.naive_bayes import GaussianNB

Y=df['Class']
X=df.drop(['Class'],axis=1)
train_x, test_x, train_y, test_y = train_test_split(X,Y,test_size=0.3)


Naive=naive_bayes.GaussianNB()
Naive.fit(train_x,train_y)
pred_y=Naive.predict(test_x)

print("Accuracy:",metrics.accuracy_score(test_y, pred_y))
print(metrics.confusion_matrix(test_y,pred_y))
```

```python
from sklearn.model_selection import cross_val_score
res2=(cross_val_score(Naive,train_x,train_y,cv=10)) # prints scores of classifier for each run of the cross validation
print("** 10-fold cross validation using NB**")
for i in range(len(res2)):
    print('Accuracy in run ',(i+1)," :: " ,res2[i])
print("\n\nMean Accuracy==>\t",np.mean(res2))
```
